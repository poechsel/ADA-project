```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib
%matplotlib inline
from igraph import *

# Import and start a Spark session.
from pyspark.sql import *
from pyspark.sql.functions import *
spark = SparkSession.builder.getOrCreate()
import dateutil.parser
import warnings; warnings.simplefilter('ignore')
```

```python
def extract_username_from_link(link):
    return link.split('/')[3]
```

```python
DATA_PATH = 'balancetonporc.json'
#data = spark.read.json(DATA_PATH)
data = spark.read.json(DATA_PATH)
```

```python
print(data.columns)
pd_data = data.toPandas()
```

```python
pd_data.head(5)
```

In order to determine the most influential twitter user for a given hashtag, we use the Pagerank algorithm. First, we build a graph using the following rule : There is an edge A -> B if A has retweeted a tweet with the given hashtag from B. Hence, we consider the most important vertices, the ones with the highest outdegree. In this case, we are performing a reverse pagerank, so the most influential nodes are the ones with the smallest pagerank weight

```python
graph = Graph()
```

```python
# Build edge A->B if A has retweeted B

def get_edges():
    edges = pd.DataFrame()
    retweets = pd_data[pd_data.is_quote_status == 1]
    edges['to'] = retweets.quote_url.map(extract_username_from_link)
    edges['from'] = retweets.username
    return edges
```

```python
get_edges()
```

```python
# Build a username - int map, which will be needed when building the graph

user_id = 0
map_user_id = {}
for edge in edges:
    for i in range(2):
        if edge[i] not in map_user_id.keys():
            map_user_id[edge[i]] = user_id
            user_id +=1
```

```python
def edge_to_id(e):
    return (map_user_id[e[0]],map_user_id[e[1]])
```

```python
# Builds the graph for the retweets
g = Graph(directed = True)
g.add_vertices(user_id)
edges_id = list(map(lambda e: edge_to_id(e),edges))
g.add_edges(edges_id)
```

```python
pr = g.pagerank()
```

```python
ind = pr.index(np.max(pr))
most_influential = ''

for k,v in map_user_id.items():
    if v == ind:
        most_influential = k
        
print(most_influential)
```

We now verify the above results by checking the number of tweets of the most influential user, along with its timeline, to see whether the peak of user's activity for the given hashtag corresponds to the peak of the hashtag.

```python
influential_tweets = pd_data[pd_data['username'].apply(lambda u: u==most_influential)]
```

```python
influential_tweets.head(5)
```

We can see from the above cell that the most influential user has contributed to 0.26% of the all tweets. This is a considerable amount, given that the total number of tweets is around 71500 and there are 36000 unique users that have contributed to the given hashtag 

```python
tmp = [('second',0),('minute',1),('hour',2),('day',3),('month',4),('year',5)]
time_to_num = dict(tmp)
```

```python
def countby_upto(t,df):
    df['timedate'] = df['date'] + ' ' + df['time']
    df['timedate'] = df['timedate'].apply(lambda t:dateutil.parser.parse(t))
    df1 = df.copy()
    df1 = df1.set_index('timedate')
    df1['timedate'] = list(map(lambda el: el[:-3 * time_to_num[t]],list(map(str,df1.index))))
    return df1.groupby(df1['timedate']).size().reset_index().rename(columns = {0:'count'}).set_index('timedate')
```

```python
plt.plot(np.log2(countby_upto('hour',pd_data)),'b-')
plt.plot(np.log2(countby_upto('hour',influential_tweets)),'r-')
```

From the above graph, it is hard to distinguish a relation between the most influential user's pattern of activity and the hashtag's activity. I think that these results from the relatively low nature of the hashtag, so it may take days for the user to influence the activity of the hashtag. Hence, it may be useful to explore the relation between the two time series, but now considering day as our unit.

```python
plt.plot(np.log2(countby_upto('day',pd_data)),'b-')
plt.plot(np.log2(countby_upto('day',influential_tweets)),'r-')
```

Now, it is much more clear that there exists a close relation in activity between the most influential user and the hashtag, especially in terms of anomalies (highs and lows).
